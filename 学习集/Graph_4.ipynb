## Conditional Graph(条件图)
Objectives:  
1) Implememt conditional logic to route the flow of data to different nodes;(实现条件逻辑、将数据流路由到不同的节点)
2) Use START and END nodes to manage entry and exit point explicity;（使用开始和结束节点显式的管理入口和出口点）
3) Design multiple nodes to perform different operations(addition,subtraction).（减法）
4) Create a router node to handle decision-making and control graph  flow.（创建一个路由点来处理决策并控制全局流）
