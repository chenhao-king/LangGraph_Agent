### 将代码推送到Github上


1) 创建.gitignore 文件来隐藏私人密钥

2) 在项目根目录处，新建一个文件，严格命名为.gitignore

3) 输入下面内容以及保存

```python
# 忽略环境变量文件
.env

#忽略 Python 缓存和虚拟环境
__pycache__/
*.__pyc
.ipynb_checkpoints/
```

4) 在github上新建一个仓库   
登录你的 GitHub 账号，点击右上角的 "+" 号，选择 New repository。   

Repository name（仓库名）：可以起个类似 LangGraph-Agent-Practice 的名字。   

下面的选项（如 Public/Private，Add a README file 等）全部保持默认（不要勾选）。   

点击绿色的 Create repository 按钮。   

创建成功后，页面上会显示一段带有 .git 结尾的链接（例如 https://github.com/你的用户名/LangGraph-Agent-Practice.git），复制这个链接备用。   

5) 在终端中绑定并推送代码  
回到 VS Code 或 Cursor 的终端中（确保终端当前的路径是你的项目文件夹），依次输入以下命令。

```python
# 初始化git
git init
# 解绑原作者仓库
git remote remove origin 
# 添加创建的文件
git add .
# 给这次上传打上标签（写提交信息）
git commit -m "init: langgraph agent environment and basic setup"
# 切换主分支名称为 main
git branch -M main
#绑定你刚刚在第二步复制的 GitHub 仓库地址：
git remote add origin 你刚刚复制的GitHub链接
#正式推送到 GitHub：
git push -u origin main
```
