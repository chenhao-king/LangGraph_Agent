### 并集
now we move on to another type of annotation which is union;   
in more complicated applications,it is actually useful;in fact,the makers of langchain or loanggraph used union quite extensively;   
1)it is flexible and it is easy to code and it allows for type safely;   
2)because it can provide hints to help catch(抓住、理解）incorrect(错误的) usage（使用，用法）;


```python
from typing import Union 
def square(x:Union[int,float]) -> float:
    return x * x
x = 5  # this is fine because it is an integer
x = 1.234  # this is also fine because it is a float 
x = "i am a string!"   # this will fail because it is a string 

# I created a very simple function which takes in a  value and it aquares it;
# now input x could be either an integer or float;
# Union basically says that whatever value you have can be these data types only;so this case x can only be integer or float
```

### Optional（可选的）
somethong similar to union is another type annotation which is optional;   


```python
from typing import Optional
def nice_message(name:Optional[str])-> None:
  if name is None:
      print("hey random person")
  else：
      print（f"Hi there ,{name}!"）
# in this case,"name "can be either String or None !
# it cannot be anything else;it cannot be an integer or boolean or a  float or anything ;
# it has to be either a string or a none;
```

### Any(任意值)
any is really the easiest one to understand;it literally(字面的) means this value could be anything;it could be any data structure(or any data type)


```python
from typing import Any
def print_value(x:Any)
    print(x)
print_value("I pretend to be Batman in the shower sometimes")
# this function called print value where it takes somethings and it print that;(接收并且打印出来；)
```
